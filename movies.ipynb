### Now I only want to keep the franchises that have 2 or more movies in the dataset

```python
#I am couning the number of movies in each franchise and saving it as franchise_counts
franchise_counts = movies_with_franchises['Franchise'].value_counts()
#and then i filter to keep only franchises with 2 or more movies
franchises_to_keep = franchise_counts[franchise_counts >= 2].index
#if the movie from the movie_with_franchises is in the franchises_to_keep list, it gets stored into the filtered_movies dataset
filtered_movies = movies_with_franchises[movies_with_franchises['Franchise'].isin(franchises_to_keep)]
#I sort the dataset by franchise
filtered_movies.sort_values(by='Franchise', ascending=False).reset_index()
```

### Now I want to see the franchises with highest ratings

```python
#I only want the sorted ratings, so I group the dataset by franchise and average rating. Then, I sort it from highest to lowest
sorted_ratings = filtered_movies.groupby('Franchise')['averageRating'].mean().sort_values(ascending=False)
sorted_ratings
```

```python
#then I save it as a csv
sorted_ratings.to_csv("sortedFranchisesRatings.csv")
```

### Let's see the total average of the franchises ratings

```python
#then.. I recall it and I check the median of the averageRating column, to check the average of the Ratings of all franchises
total_avg = pd.read_csv("sortedFranchisesRatings.csv")
total_avg['averageRating'].median()
```

### I want to create two columns: one for the movies that are rated above 7.5 (the average of ratings) and one for those below

```python
#Then I create two new columns for ratings above and below or equal to 7.5, by using the lambda funtion
total_avg['Above 7.5'] = total_avg['averageRating'].apply(lambda x: x if x >= 7.5 else None)
total_avg['7.5 and Below'] = total_avg['averageRating'].apply(lambda x: x if x < 7.5 else None)
#i save it as a csv
total_avg.to_csv("aboveAndBelow.csv")
```

```python
#back to the filtered_movies dataset, i keep the necessary columns
filtered_movies = filtered_movies[['averageRating',	'numVotes',	'Film',	'date',	'genres',	'Franchise']]
```

### I want to get information about the original movies and the sequels. Which is usually the higher rated?

```python
#I turn the date column into datetime
filtered_movies['date'] = pd.to_datetime(filtered_movies['date'], format='%Y')

#Sorting the dataframe by franchise and date
filtered_movies = filtered_movies.sort_values(by=['Franchise', 'date'])
#Identifying the first movie for each franchise
first_movies = filtered_movies.drop_duplicates(subset=['Franchise'], keep='first')
#grouping by franchise and anerage rating to find the highest ratinngs
highest_ratings = filtered_movies.groupby('Franchise')['averageRating'].max().reset_index()
#then, I merge the first movies and the highest rating
comparison_df = pd.merge(first_movies, highest_ratings, on='Franchise', suffixes=('', '_max'))

#Checking to see if the first movie has the highest rating
comparison_df['is_first_highest'] = comparison_df['averageRating'] == comparison_df['averageRating_max']

#I diplay the dataframe
comparison_df[['Franchise', 'Film', 'date', 'averageRating', 'averageRating_max', 'is_first_highest']]
```

```python
#I create a def funtion so see if the movie is a sequel or not
def is_sequel(film, franchise):
    #Finding the original film in the franchise by selecting the film with the earliest release date
    original =filtered_movies[(filtered_movies['Franchise'] == franchise) & (filtered_movies['date'] == filtered_movies[filtered_movies['Franchise'] == franchise]['date'].min())]['Film'].values[0]
     # Return True if the film is not the original, indicating it is a sequel
    return film != original

# Applying the is_sequel function to each row in the filtered_movies DataFrame
# Creating a new column 'is_sequel' that indicates if the film is a sequel (True) or not (False)
filtered_movies['is_sequel'] = filtered_movies.apply(lambda row: is_sequel(row['Film'], row['Franchise']), axis=1)

#I want to seperate the dataset between originals and sequels
originals = filtered_movies[filtered_movies['is_sequel'] == False]
sequels = filtered_movies[filtered_movies['is_sequel'] == True]

# Calculating the average rating for original and sequel films in each franchise
originals_avg = originals.groupby('Franchise')['averageRating'].mean()
sequels_avg = sequels.groupby('Franchise')['averageRating'].mean()

# Combining the average ratings of originals and sequels into a single dataframe for comparison
comparison = pd.DataFrame({'Originals': originals_avg, 'Sequels': sequels_avg}).dropna()

#Displaying the dataset
comparison
```
