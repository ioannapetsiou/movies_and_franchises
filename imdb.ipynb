### In this notebook I will be using data from Imdb movies

```python
#importing libraries
import pandas as pd
import numpy as np
import requests 
import matplotlib.pyplot as plt

#pd.set_option("display.max_rows", None)
#pd.set_option("display.max_columns", None)
```

```python
#I have a file in my computer called "title.ratings.tsv.gz" and I am calling it as a dataframe
#this file is "tab seperated" and I am changing that so that it works as a csv
df = pd.read_csv("title.ratings.tsv.gz",
                 sep="\t",
                 encoding="latin-1",
                 na_values=[0])
df
```

```python
#same as the cell above - I am calling a dataset and displaying it
df2 = pd.read_csv("title.basics.tsv.gz",
                 sep="\t",
                 encoding="latin-1",
                 na_values=[0])
df2
```

```python
#only keeping the rows were the titleType is a movie, and displaying the changed dataset
df2 = df2[df2['titleType'].str.lower() == "movie"]
df2
```

```python
#merging the 2 datasets from before into one on the "tconst" column, that seems to be like a code for each movie
merged_df = pd.merge(df, df2, on='tconst', how='inner')
merged_df
```

```python
#sorting the new merged dataframe by the number of Votes so that i only get popular movies. Then i reset the index
merged_df = merged_df.sort_values(by='numVotes', ascending=False).reset_index(drop=True)
merged_df
```

```python
#I only want to keep the first 500 rows of the sorted dataframe
merged_df = merged_df.iloc[:500]
merged_df
```

```python
#I only want to keep specific columns, so once again I resave my dataframe with what I want to keep
merged_df = merged_df[['averageRating','numVotes','originalTitle','startYear', 'genres']]
merged_df
```

```python
#now I sort my dataframe again by the ratings of each movie
merged_df = merged_df.sort_values(by='averageRating', ascending=False).reset_index(drop=True)
merged_df
```

```python
#I rename the title and year column to match a dataset I will be using in an other motebook
merged_df = merged_df.rename(columns={"originalTitle": "Film", "startYear": "date"})
```
